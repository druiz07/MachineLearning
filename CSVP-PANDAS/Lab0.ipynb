```python
import random
from scipy.integrate import quad
import numpy as np

def cuadrado(x):
    return x*x

def integra_mc(fun, a, b, m,num_puntos=10000):
   
    puntos=0
    for i in range (num_puntos):
        x=random.randint(a,b)
        y= random.randint(0,m)
        
        if (y < fun(x)):
            puntos=puntos+1
    
    
    return (puntos/num_puntos)*(b-a)*m
    

print(integra_mc(cuadrado,-1,1,1))
print(quad(cuadrado,-1,1))
 
def integra_mc_vectorizada(fun, a, b, m, num_puntos=10000):
    # Generar num_puntos valores aleatorios en el intervalo [a, b] para x
    x = np.random.uniform(a, b, num_puntos) #Se genera un vector de puntos de x
    
    # Generar num_puntos valores aleatorios en el intervalo [0, m] para y
    y = np.random.uniform(0, m, num_puntos) #Se genera un vector de puntos de y
    
    # Calcular la cantidad de puntos que estÃ¡n por debajo de la funciÃ³n
    puntos_bajo_funcion = y < fun(x)  #Se guardan de forma vectorizada (ARRAY DE BOOL) en puntos bajo funcion
    #todos los puntos que cumplan y < fun(x) (Aplicado a cada punto ya que x e y son vectores )
    
    # Calcular el resultado de la integral
    #Usando np.sum calculamos a partir de los puntos bajo funcion( estan a true en el array puntos bajo funcion )
    # La funciÃ³n np.sum suma todos los valores True en este array booleano, 
    #lo que proporciona la cantidad de puntos que cumplen esta condiciÃ³n.
    resultado = (np.sum(puntos_bajo_funcion) / num_puntos)* (b - a) * m
    
    return resultado

print(integra_mc_vectorizada(cuadrado,-1,1,1))
print(quad(cuadrado,-1,1))
```
